First, I'll start by creating a dataframe for total net emissions, and then one dataframe for each sector. This will allow me to manipulate data in an easier fashion.

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from aquarel import load_theme

df = pd.read_csv('data/Greenlands total emission of greenhousegasses by greenhousegas, sector and time.csv')
```

```python
df.head()
```

Let's calculate some statistics: mean and st dev for total emissions, calculate the first and third quantiles and interquantile range to check for possible outliers. Also min and max values and their corresponding year.

```python
#create a dataframe for the total value of net greenhouse gas emissions throughout the years
total_df = df.query('sector == "Total (net emissions)"').drop(columns=['greenhousegas'])

#explore some statistics

#get the mean value for emissions over the period between 1990-2020
mean_emissions = total_df['Greenlands total emission of greenhousegasses'].mean()
std_dev_emissions = total_df['Greenlands total emission of greenhousegasses'].std()


#Because the variation is relatively big, lets calculate the interquartile range and find the outliers for this dataset

# Compute the first and third quantiles and IQR of emissions_by_country
q1 = np.quantile(total_df['Greenlands total emission of greenhousegasses'], 0.25)
q3 = np.quantile(total_df['Greenlands total emission of greenhousegasses'], 0.75)
iqr = q3 - q1

# Calculate the lower and upper cutoffs for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Find outliers
outliers = total_df[(total_df['Greenlands total emission of greenhousegasses'] < lower_bound) | (total_df['Greenlands total emission of greenhousegasses'] > upper_bound)]


# Create a box plot
plt.boxplot(total_df['Greenlands total emission of greenhousegasses'])

# Plot the interquartile range (IQR)
plt.axhline(y=q1, color='aquamarine', linestyle='--', label='Q1')
plt.axhline(y=q3, color='aquamarine', linestyle='--', label='Q3')
plt.axhline(y=q1 - 1.5 * iqr, color='salmon', linestyle='--', label='Lower Bound')
plt.axhline(y=q3 + 1.5 * iqr, color='salmon', linestyle='--', label='Upper Bound')

# Add legend
plt.legend(loc="upper right", fontsize="small", bbox_to_anchor=(1.4, 1.))

plt.savefig('Interquartile range total emissions.jpg', bbox_inches='tight')

# Show plot
plt.show()

#Find the index of the maximum value
max_index = total_df['Greenlands total emission of greenhousegasses'].idxmax()

# Get the corresponding year value using the index
max_year = total_df.loc[max_index, 'time']

# Get the maximum value
max_value = total_df['Greenlands total emission of greenhousegasses'].max()

#Find the index of the minimum value
min_index = total_df['Greenlands total emission of greenhousegasses'].idxmin()

# Get the corresponding year value using the index
min_year = total_df.loc[min_index, 'time']

# Get the minimum value
min_value = total_df['Greenlands total emission of greenhousegasses'].min()
```

```python
print(f'The average emissions level over the specified period is {mean_emissions} units.\
There is a moderate amount of variability in the emissions data, with emissions values typically deviating \
from the average by approximately {std_dev_emissions} units.')

print(f'The maximum value is {max_value} carbon dioxide equivalents (CO2e), as established by the United Nations Framework Convention on Climate Change (UNFCCC), corresponding to the year of {max_year}')

print(f'The minimum value is {min_value} carbon dioxide equivalents (CO2e), corresponding to the year of {min_year}')


print("Lower bound for outliers:", lower_bound)

print("Upper bound for outliers:", upper_bound)

print("Outliers:", outliers)

print('There are not outliers in this dataset')
```

```python
#aquarel library theme
theme = load_theme("arctic_dark")

total_df.plot('time','Greenlands total emission of greenhousegasses', label='total emissions')

# Add a line for the mean value
plt.axhline(mean_emissions, color='ivory', linestyle='--', label='Mean Emissions')

#create two flags to annotate max and min values of emissions in the graph
# Annotate the maximum value
plt.annotate(f'Max Emission: {max_value}', xy=(max_year, max_value),
             xytext=(max_year, max_value + 5), arrowprops=dict(facecolor='salmon', shrink=0.05))

# Annotate the minimum value
plt.annotate(f'Min Emission: {min_value}', xy=(min_year, min_value),
             xytext=(min_year, min_value - 5), arrowprops=dict(facecolor='aquamarine', shrink=0.05))

plt.title('Total emissions over time in Greenland')

plt.xlabel('Years')

plt.ylabel('Emissions (Co2e)')

plt.legend(loc="upper right", fontsize="small", bbox_to_anchor=(1.2, 1.2))

plt.grid(True)

theme.apply()

plt.savefig('Total emissions over time in Greenland.jpg', bbox_inches='tight')

plt.show()
```

Year-on-Year Changes: Analyze the changes in total emissions over multiple years to identify trends and patterns. The emissions change column shows the difference in emissions between each year and the previous year. Positive values indicate an increase in emissions compared to the previous year, while negative values indicate a decrease.

```python
# Calculate year-on-year changes in total emissions
total_df['emissions_change'] = total_df['Greenlands total emission of greenhousegasses'].diff()
total_df.head()
```

```python
total_df.plot('time', 'emissions_change', label = 'change in emissions')

# Flag the year corresponding to the maximum emission
plt.axvline(x=max_year, color='salmon', linestyle='--', label=f'Max Emission Year: {max_year}')

# Flag the year corresponding to the minimum emission
plt.axvline(x=min_year, color='aquamarine', linestyle='--', label=f'Min Emission Year: {min_year}')


plt.title('Rate of change in total emissions over time')

plt.xlabel('Years')

plt.ylabel('Change in emissions')

plt.legend('yearly differences in emissions')

plt.legend(loc="upper right", fontsize="small", bbox_to_anchor=(1.3, 1.3))

theme.apply()

plt.savefig('Rate of change in total emissions over time.jpg', bbox_inches='tight')

plt.show()
```

Let's preform a regression analysis using polynomial fitting on the total greenhouse gas emissions in Greenland from 1990 to 2020 to understand if there are any emission trends and if they are significant. I chose a polynomial transformation of degree 4 (quartic polynomial function), since I could see it would probably fit my data better and I would probably be underfitting If I chose a linear polynomial (polynomial transformation of degree 1) in this situation.
Note: Although this model allows for a nonlinear relationship between the independent and dependent variables, polynomial regression is still considered linear regression since it is linear in the regression coefficients.

```python
# Regression analysis (using numpy.polyfit)
coefficients = np.polyfit(total_df['time'], total_df['Greenlands total emission of greenhousegasses'], 4) #we're using a quartic polynomial function. If it was linear we would use 1

#calculate the slope of regression line
slope = coefficients[0]

# Calculate predicted values
predicted_values = np.polyval(coefficients, total_df['time'])

# Calculate residuals
residuals = total_df['Greenlands total emission of greenhousegasses'] - predicted_values

# Compute sum of squares
SST = np.sum((total_df['Greenlands total emission of greenhousegasses'] - np.mean(total_df['Greenlands total emission of greenhousegasses'])) ** 2)
SSR = np.sum((predicted_values - np.mean(total_df['Greenlands total emission of greenhousegasses'])) ** 2)
SSE = np.sum(residuals ** 2)

# Calculate R-squared
R_squared = SSR/SST

# Calculate RMSE
RMSE = np.sqrt(SSE / len(total_df))

print("R-squared:", R_squared)
print("RMSE:", RMSE)

print("Slope of Regression Line:", slope)
```

Approximately 61.6% of the variance in the dependent variable (emissions) is explained by the independent variables (time). This suggests a moderately strong relationship between time and emissions in your model. Overall, an R-squared value of 0.616 suggests that my model captures a substantial portion of the variance in emissions, but the relatively high RMSE indicates that there may be considerable variability in the model's predictions.

```python
# Plot residuals against predicted values
plt.figure(figsize=(8, 6))
plt.scatter(predicted_values, residuals, alpha=0.5)
plt.axhline(y=0, color='salmon', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs. Predicted Values Plot')
plt.grid(True)

plt.savefig('Residuals vs predicted values.jpg', bbox_inches='tight')

plt.show()
```

```python
# Plotting the data points
plt.figure(figsize=(10, 6))
plt.scatter(total_df['time'], total_df['Greenlands total emission of greenhousegasses'], label='total emissions')

# Generate x-values for plotting (assuming time values range from min to max)
x_values = np.linspace(total_df['time'].min(), total_df['time'].max(), 100)

# Calculate corresponding y-values using the polynomial equation
y_values = np.polyval(coefficients, x_values)

# Plotting the regression line
plt.plot(x_values, y_values, color='salmon', label='Regression Line (predicted emissions)')

plt.xlabel('Year')
plt.ylabel('Total net emissions (Co2e)')
plt.title('Regression Line of Total Net Emissions in Greenland Over the Years')
plt.legend(bbox_to_anchor=(1.2, 1.2))
plt.grid(True)

plt.savefig('Regression Line of Total Net Emissions Over the Years.jpg', bbox_inches='tight')

plt.show()
```

```python
pred_df = pd.DataFrame({'Real Values': total_df['Greenlands total emission of greenhousegasses'], 'Predicted values': predicted_values})

pred_df.head()
```

```python
# Extend the time array by adding 50 years
extended_time = np.arange(total_df['time'].min(), total_df['time'].max() + 51)

# Calculate predicted emissions for the extended time range
predicted_emissions_next_50_years = np.polyval(coefficients, extended_time)

# Plot the original data and predicted values
plt.figure(figsize=(10, 6))

plt.scatter(total_df['time'], total_df['Greenlands total emission of greenhousegasses'], label='Original Data')

plt.plot(extended_time, predicted_emissions_next_50_years, color='salmon', label='Predicted Values')

plt.xlabel('Year')

plt.ylabel('Emissions Co2e')

plt.title('Real and predicted values for the next 50 Years')

plt.legend(bbox_to_anchor=(1.2, 1.2))

plt.savefig('predicted emissions for the next 50 years.jpg', bbox_inches='tight')

plt.show()
```
